```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
import mapclassify as mc
import geoplot as gplt
# from matplotlib_scalebar.scalebar import ScaleBar
```

```python
plt.rcParams.update({'font.size':27})
```

```python
mapa_para = gpd.read_file('mapas/PA_Mesorregioes_2021.zip') # Ler arquivo

mapa_para['NM_MESO'] = ['Lower Amazon',
                        'Marajó',
                        'MA',
                        'Northeastern',
                        'Southwest',
                        'Southeast']

dat = pd.read_csv('mapas/mesoregiao_pa_2.csv')
dat['CD_MESO'] = dat['CD_MESO'].astype(str)
```

```python
mapa_meso_2019 = mapa_para.merge(# Função para juntar geopandas
    dat.groupby(['CD_MESO'])[['First day of the 2019 exam', 'Second day of the 2019 exam']]
    .mean()
    .reset_index()
)

mapa_meso_2020 = mapa_para.merge(# Função para juntar geopandas
    dat.groupby(['CD_MESO'])[['First day of the 2020 exam', 'Second day of the 2020 exam']]
    .mean()
    .reset_index()
)

mapa_meso_2021 = mapa_para.merge(# Função para juntar geopandas
    dat.groupby(['CD_MESO'])[['First day of the 2021 exam', 'Second day of the 2021 exam']]
    .mean()
    .reset_index()
)

mapa_meso_2022 = mapa_para.merge(# Função para juntar geopandas
    dat.groupby(['CD_MESO'])[['First day of the 2022 exam', 'Second day of the 2022 exam']]
    .mean()
    .reset_index()
)
```

```python
dat_update = pd.read_csv('mapas/mesoregiao_pa_2 - Edição.csv', sep=',')
df = dat_update[dat_update['Test day'] == 'Second day ']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
import geoplot as gplt
import pandas as pd
import matplotlib.colors as mcolors

# Tema e estilo
sns.set_theme(style="white", palette="pastel")
sns.despine()

# Substituir "Metropolitan Area" por "MA" no DataFrame
df['Meso-Regions'] = df['Meso-Regions'].replace("Metropolitan Area", "MA")

# Configurar a figura com subplots (um para o mapa e outro para o gráfico de barras combinado)
fig, axes = plt.subplots(1, 2, figsize=(20, 10), gridspec_kw={'width_ratios': [1.5, 1]})

# Função para adicionar a bússola
def add_compass(ax, x=0.0, y=0.9):
    ax.annotate('N', xy=(x, y), xytext=(x, y-0.1),
                arrowprops=dict(facecolor='black', width=5, headwidth=15),
                ha='center', va='center', fontsize=24, xycoords=ax.transAxes)

# Função para criar um dicionário de cores baseado na coluna NM_MESO
def get_color_mapping(gdf, cmap_name='Set2'):
    gdf['NM_MESO'] = gdf['NM_MESO'].replace("Metropolitan Area", "MA")
    unique_regions = gdf['NM_MESO'].unique()
    cmap = plt.cm.get_cmap(cmap_name, len(unique_regions))
    color_dict = {region: mcolors.rgb2hex(cmap(i)) for i, region in enumerate(sorted(unique_regions))}
    return color_dict

# Selecionar um dos mapas para exibir
mapa_mesorregioes = mapa_meso_2020  # Selecionar o mapa do ano de 2020, por exemplo
color_mapping = get_color_mapping(mapa_mesorregioes, cmap_name='Set2')

# Mapa das Mesorregiões sem legenda
gplt.choropleth(mapa_mesorregioes.replace("Metropolitan Area", "MA"), hue='NM_MESO', cmap='Set2', legend=False, ax=axes[0])
axes[0].set_title("Map of Pará's Mesoregions", fontsize=24)
mapa_mesorregioes.apply(lambda x: axes[0].annotate(text=x['NM_MESO'].replace("Metropolitan Area", "MA"), fontsize=24,
                        xy=x.geometry.centroid.coords[0], ha='center'), axis=1)
add_compass(axes[0])

# Gráfico de barras combinado com as mesmas cores do mapa
df['Meso-Regions'] = df['Meso-Regions'].replace("Metropolitan Area", "MA")

# Plotar o gráfico de barras com as cores consistentes e espaçamento entre os valores de hue
sns.barplot(data=df, x='Year', y='Abstention rate', hue='Meso-Regions', palette=color_mapping, ax=axes[1], dodge=True)
axes[1].axhline(y=20, color='red', linestyle='--', linewidth=1.5)  # Linha de referência horizontal no valor 10
axes[1].set_title("Abstention rate", fontsize=24)
axes[1].set_xlabel("Year", fontsize=24)
axes[1].set_ylabel("Abstention (%)", fontsize=24)

# Remover a legenda do gráfico de barras, mantendo a legenda geral intacta
if axes[1].get_legend() is not None:
    axes[1].get_legend().remove()

# Ajuste do tamanho da fonte dos rótulos dos ticks
axes[1].tick_params(axis='x', labelsize=24)
axes[1].tick_params(axis='y', labelsize=24)

# Criar uma legenda compartilhada na parte inferior da figura
handles, labels = axes[1].get_legend_handles_labels()
fig.legend(handles, labels, title="Mesoregions", fontsize=22, title_fontsize=22, loc="lower center", ncol=len(labels))

# Ajuste final e salvamento da figura
plt.tight_layout(rect=[0, 0.1, 1, 1])  # Ajuste para espaço da legenda inferior
plt.savefig("map_and_abstention_rate.png", format="png", dpi=300)  # Salvar como TIFF com 300 DPI
plt.show()
```
