```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(14,7)})
```

```python
from sktime.utils.plotting import plot_series
```

```python
y = pd.read_csv('data/serie_historica_inscritos_enem_brasil.csv', parse_dates=['date'], index_col='date')
del y['local']

z = y.iloc[:-1, :]
```

```python
# plotting for visualization
plot_series(y)
```

```python
# plotting for visualization
plot_series(z)
```

```python
from sktime.forecasting.base import ForecastingHorizon
from sktime.forecasting.naive import NaiveForecaster
from sktime.forecasting.arima import ARIMA, AutoARIMA
```

```python
coverage = 0.95
# step 2: specifying forecasting horizon
fh = list(np.arange(1, 5))
# step 3: specifying the forecasting algorithm
forecaster = AutoARIMA(seasonal=False, sp=3)
# step 4: fitting the forecaster
forecaster.fit(y.values)
```

```python
forecaster.summary()
```

```python
# step 5: querying predictions
y_pred = forecaster.predict(fh=fh)
```

```python
y_pred
```

```python
array = list()

for i in range(len(fh)):
    present =  '2022-12-31'
    new_date = pd.to_datetime(present)+pd.DateOffset(years= i)
    array.append(new_date)
    
pred = pd.DataFrame()
pred['date'] = array
pred['inscritos'] = y_pred 

pred.set_index('date', inplace=True)
```

```python
# optional: plotting predictions and past data
plot_series(y, pred, labels=["y", "pred"])
```
